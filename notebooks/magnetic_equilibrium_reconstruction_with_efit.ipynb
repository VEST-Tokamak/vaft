# Magnetic Equilibrium Reconstruction with EFIT

## Overview

This notebook is a first-draft shell for documenting EFIT-based magnetic equilibrium reconstruction in the VAFT pipeline. It focuses on magnetic diagnostics, eddy-current inputs, PF coil information, and physical constraints required for equilibrium reconstruction.

## Objectives

- Define the EFIT reconstruction inputs produced by upstream VAFT stages.
- Document magnetic diagnostics, eddy-current estimates, and PF coil current requirements.
- Organize physical constraints and reconstruction configuration parameters.
- Specify expected EFIT outputs for downstream equilibrium and stability analysis.

## Expected Inputs

- Processed magnetic diagnostics with calibration metadata.
- PF active coil currents and PF passive eddy-current estimates.
- Plasma current, pressure constraints, and boundary or limiter information.
- EFIT configuration files, time slices, and physical constraint definitions.

## Expected Outputs

- EFIT reconstruction results for selected shots and time slices.
- Equilibrium files, flux maps, boundary descriptions, and residual summaries.
- Diagnostics-fit quality metrics and constraint consistency checks.
- Inputs for MHD equilibrium interpretation and stability workflows.

## Pipeline Context

This notebook documents the equilibrium reconstruction stage of the Snakemake-based VAFT pipeline. It should eventually consume processed diagnostics and eddy-current products, then produce versioned EFIT equilibria for MHD analysis, DCON, RDCON, and GPEC workflows.
