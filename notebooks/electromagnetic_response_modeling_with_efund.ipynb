# Electromagnetic Response Modeling with EFUND

## Overview

This notebook is a first-draft shell for documenting electromagnetic response modeling with EFUND in the VAFT pipeline. It covers magnetic diagnostics, wall geometry, PF active and PF passive geometry, material resistivity, Green's function-based response calculation, electromagnetic coupling matrix construction, and comparison with EFUND module and table structure.

## Objectives

- Organize geometry inputs for magnetic diagnostics, wall structures, PF active coils, and PF passive conductors.
- Document material resistivity and circuit parameter requirements.
- Outline Green's function-based response calculations.
- Define electromagnetic coupling matrix construction and validation against EFUND outputs.

## Expected Inputs

- Magnetic diagnostics geometry and channel metadata.
- Wall, PF active coil, and PF passive conductor geometry definitions.
- Material resistivity, resistance, inductance, and mutual-coupling parameters.
- EFUND module outputs, database tables, or reference matrix structures.

## Expected Outputs

- Green's function response products for diagnostics and conductors.
- Electromagnetic coupling matrices with documented indexing and units.
- Comparison tables between VAFT calculations and EFUND module outputs.
- Open compatibility notes for EFUND table structure and pipeline integration.

## Pipeline Context

This notebook documents the electromagnetic response layer that supports eddy-current modeling and magnetic equilibrium reconstruction. In the Snakemake pipeline, these products should become reusable geometry and response-matrix intermediates.
